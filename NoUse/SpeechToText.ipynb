```python
# Extracted first 10 seconds from the first mp3, and saved it as output_1.mp3

import subprocess

from whisper import transcribe

subprocess.run([
    "ffmpeg",
    "-i", "/Users/frolt/Git Projects/Data_Science/DS_ML_Projects/RAGbasedAI_assistant/audios/Lecture_1.mp3",
    "-t", "10",
    "output_1.mp3"
])
```

```python
import whisper
import json

model = whisper.load_model("small")     # Loads a Whisper model named "large-v3" into memory and assigns it to model.

result = model.transcribe("output_1.mp3"                # called the transcribe method from model class.
                          , language="en"          # passed arguments: audio file path, language code
                          , task = "transcribe"
                          , word_timestamps=False )     # this will give timestamps for each word
             # prints the transcribed text from the audio file
with open ("output1.json", "w", encoding="utf-8") as f:       # created a new JSON file to save the transcription

  json.dump(result, f)                                        # dumped the transcription result into the JSON file

chunks = []                        # initialized an empty list to store text chunks
for segment in result['segments']:        # iterated over each segment in the transcription result
    chunks.append({"start": segment["start"], "end": segment["end"], "text": segment["text"]})

    print(chunks)                     # printed the list of text chunks with their start and end times

json.dump(chunks, open("output1.json", "w", encoding="utf-8"))


  # --------code explanation--------

  # Imports the Whisper library (the speech-to-text model wrapper) so you can load models and transcribe audio.
  # Imports Pythonâ€™s JSON module for reading/writing JSON data.
```
